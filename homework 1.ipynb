```python
import cv2
import numpy as np

drawing = False # true if mouse is pressed
mode = True # if True, draw rectangle. Press 'm' to toggle to curve
ix,iy = -1,-1


def nothing(x):
    pass

# mouse callback function
def draw_rectangle(event,x,y,flags,param,):
    text_color = (255, 255, 255)
    global ix,iy,drawing,img
    
    if event == cv2.EVENT_LBUTTONDOWN:
        img[:50,:200,:]=0
        drawing = True
        ix,iy = x,y
        cv2.rectangle(img,(ix,iy),(x,y),(0,0,255),1)
        

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            img_copy = img.copy()
            cv2.rectangle(img_copy,(ix,iy),(x,y),(0,0,255),1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.putText(img, f'first: ({ix}, {iy})', (40, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1)
        cv2.putText(img, f'last: ({x}, {y})', (40, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1)
        cv2.rectangle(img,(ix,iy),(x,y),(0,0,255),1)

        roi = img[iy:y,ix:x] # B,G값을 0으로하여 R값만 드러나도록
        roi[:, :, 0] = 0
        roi[:, :, 1] = 0
           

img = cv2.imread('bg.jpg')
cv2.namedWindow('image')
# 이미지 창 생성 및 마우스 이벤트 콜백 설정
cv2.setMouseCallback('image',draw_rectangle)

cv2.createTrackbar('Value','image',0,255,nothing)


while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

    r = cv2.getTrackbarPos('Value','image')
   



cv2.destroyAllWindows()
```
